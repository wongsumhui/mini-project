# Question 3
Write codes that performs the following:
Load the video “img_pexels.mp4” into the Python environment, resize it and display the videos with smaller frames (The frames can be of any size, as long as it is smaller). You can specify an arbitrary frame rate.
Save it as a separate files: “smaller_img_pexels.avi” or "smaller_img_pexels.mp4"

```python
import cv2

# Input and output file paths
input_video_path = 'img_pexels.mp4'
output_video_path = 'smaller_img_pexels.avi'  # or .mp4, based on preference

# Open the input video file
cap = cv2.VideoCapture(input_video_path)

# Check if the video opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get the video's original frame width, height, and frame rate
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define the new dimensions for resizing (e.g., smaller by half)
new_width = int(frame_width / 2)
new_height = int(frame_height / 2)

# Create VideoWriter object to save the output video
out = cv2.VideoWriter(output_video_path, cv2.VideoWriter_fourcc(*'DIVX'), fps, (new_width, new_height))

while True:
    # Read a frame from the video
    ret, frame = cap.read()

    if not ret:
        break

    # Resize the frame
    resized_frame = cv2.resize(frame, (new_width, new_height))

    # Display the resized frame
    cv2.imshow('Resized Video', resized_frame)
    cv2.waitKey(25)  # Adjust the wait time (ms) as needed for the desired frame rate

    # Write the resized frame to the output video file
    out.write(resized_frame)

# Release the VideoCapture and VideoWriter objects
cap.release()
out.release()

# Close all OpenCV windows
cv2.destroyAllWindows()
```

#Question 4
Enlarge the image "dog.jfif" by using different techniques:
- Linear interpolation
- Cubic interpolation
- Nearest neighbor interpolation.

```python
from IPython.display import Image
```

```python
import timeit

# A. Linear interpolation
img = cv.imread("images/images/dog.jfif")

linear_time = timeit.timeit(lambda: cv.resize(img, None, fx=1.5, fy=1.5, interpolation=cv.INTER_LINEAR), number=100)
img_linear = cv.resize(img, None, fx=1.5, fy=1.5, interpolation=cv.INTER_LINEAR)

cv.imshow('linear', img_linear)
cv.waitKey(0)
cv.destroyAllWindows()
```

# B. Cubic interpolation

```python
cubic_time = timeit.timeit(lambda: cv.resize(img, None, fx=1.5, fy=1.5, interpolation=cv.INTER_CUBIC), number=100)
img_cubic = cv.resize(img, None, fx=1.5, fy=1.5, interpolation=cv.INTER_CUBIC)

cv.imshow('Cubic Interpolation', img_cubic)
cv.waitKey(0)
cv.destroyAllWindows()
```

# C. Nearest neighbor interpolation.

```python
img = cv.imread('images/images/dog.jfif')
nearest_time = timeit.timeit(lambda: cv.resize(img, None, fx=1.5, fy=1.5, interpolation=cv.INTER_NEAREST), number=100)
img_nearest = cv.resize(img, None, fx=1.5, fy=1.5, interpolation=cv.INTER_NEAREST)

cv.imshow('Nearest Neighbor Interpolation', img_nearest)
cv.waitKey(0)
cv.destroyAllWindows()
```
